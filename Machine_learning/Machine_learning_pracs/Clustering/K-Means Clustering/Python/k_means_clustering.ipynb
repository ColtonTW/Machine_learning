# K-Means Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample data (replace X with your data)
X = np.random.rand(100, 2)

# Perform Elbow Method with reduced dataset and iterations
wcss = []
max_clusters = 10  # Maximum number of clusters to try
sample_size = 1000  # Size of the dataset to use for Elbow Method (adjust as needed)

# Reduce dataset size to mitigate memory issues
X_sample = X[:sample_size]

for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=3, max_iter=100)
    kmeans.fit(X_sample)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.plot(range(1, max_clusters + 1), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

1. kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42):

2. KMeans: This is a class from the scikit-learn library used for performing KMeans clustering, a popular clustering algorithm.

3. n_clusters=5: This parameter specifies the number of clusters to form, which in this case is set to 5. KMeans will try to partition the data into 5 distinct clusters.

4. init='k-means++': This parameter specifies the method for initializing the centroids (cluster centers). 'k-means++' is a method that initializes the centroids in a smart way to speed up convergence.

5. random_state=42: This parameter is used to initialize the random number generator. Setting random_state to a specific value ensures that the results are reproducible.

6. y_kmeans = kmeans.fit_predict(X):

7. fit_predict(X): This method is used to fit the KMeans model to the data X and then predict the cluster labels for each sample in X. It is a convenience method that combines fitting the model (fit) and predicting the labels (predict) into a single step.

8. X: This is the input data, typically a numpy array or pandas DataFrame, containing the features used for clustering.

## Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
